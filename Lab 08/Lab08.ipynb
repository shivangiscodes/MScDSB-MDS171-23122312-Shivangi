```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    def print_stack(self):
        if not self.is_empty():
            print("Stack:", self.items)
        else:
            print("Stack is empty")

    def size(self):
        return len(self.items)

    def top(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return "Stack is empty"

    def is_empty(self):
        return len(self.items) == 0

def main():
    my_stack = Stack()

    while True:
        print("\nStack Menu:")
        print("1. Push Item")
        print("2. Pop Item")
        print("3. Print Items in Stack")
        print("4. Size of Stack")
        print("5. Top Item in Stack")
        print("6. Check if Stack is Empty")
        print("7. Quit")

        choice = input("Enter your choice: ")

        if choice == "1":
            item = input("Enter item to push: ")
            my_stack.push(item)
            print(item, "pushed onto the stack.")
        elif choice == "2":
            popped_item = my_stack.pop()
            print("Popped item:", popped_item)
        elif choice == "3":
            my_stack.print_stack()
        elif choice == "4":
            print("Stack size:", my_stack.size())
        elif choice == "5":
            top_item = my_stack.top()
            print("Top item:", top_item)
        elif choice == "6":
            if my_stack.is_empty():
                print("Stack is empty.")
            else:
                print("Stack is not empty.")
        elif choice == "7":
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
