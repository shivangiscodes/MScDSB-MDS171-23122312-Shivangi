```python
import random
Item=['Apple','Banana','Juice','Soap','Knife','Jam','Pickle','Cheese','Bread','Butter']
Quantity=[2,3,5,6,7,6,4,9,7,6]
Price=[20,40,50,90,80,60,70,30,100,10]
r1=random.randrange(0,len(Item))
r2=random.randint(0,len(Quantity))
r3=random.randint(0,len(Price))
string_val = Item[r1]+','+str(Quantity[r2])+','+str(Price[r3])
print(string_val)

random_data = [Item[r1], r2, Price[r3]]
random_data = f'{Item[r1]},{Quantity[r2]},{Price[r3]}\n'
with open('23122132.csv', 'a') as csvfile:
    csvfile.write(random_data)
```

```python
with open('23122132.csv','r') as csvfile:
    total_cost = 0
    for line in csvfile:
        values = line.strip().split(',')
        if len(values)>=2:
            quantity=int(values[1])
            price=float(values[2])
            row_total=quantity*price
            total_cost=total_cost+row_total
            print(f'Total cost for {values[0]}: {row_total}')
    print(f'Overall Total Cost: {total_cost}')
```
